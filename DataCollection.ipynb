```python
# Define Arduino port
import serial
import re
from datetime import datetime

arduino_port = "COM3" #serial port of Arduino
baud = 115200 #arduino uno runs at 9600 baud
fileName="PulseData7.csv" #name of the CSV file generated
```

```python
# Connect to the port
ser = serial.Serial(arduino_port, baud)
print("Connected to Arduino port:" + arduino_port)
file = open(fileName, "w")
print("Created file")
```

```python
import csv
samples = 5000 #how many samples to collect
print_labels = False
line = 0 #start at 0 because our header is 0 (not real data)
file = open(fileName, "a")
writer = csv.writer(file, lineterminator = '\n')

writer.writerow(['Time','HR'])
while line <= samples:
    # incoming = ser.read(9999)
    # if len(incoming) > 0:
    if print_labels:
            print("Line " + str(line) + ": writing...")
    getData=str(ser.readline())
    now = datetime.now()

    current_time = now.strftime("%H:%M:%S")
    
    
    data=getData[1:]
    # using re
    # Extract digit string 
    res = re.sub("\D", "", data)
    Heartrate = res
    print(current_time + " " + Heartrate)
    

    #file = open(fileName, "a")
    writer = csv.writer(file, lineterminator = '\n')

    writer.writerow([current_time, Heartrate])

    line = line+1

print("Data collection complete!")
```
